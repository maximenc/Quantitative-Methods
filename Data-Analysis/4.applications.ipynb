### **17. Average and Median ESG Scores by Sector**
Map the `GIC Sectors` column to the corresponding industry names using the following dictionary. Then, calculate the average and median ESG scores for each sector. Present the results in a table.

Hint: Use `groupby()`.

```python
# Dictionary for the mappings
industry_mapping = {
    '10': 'Energy',
    '15': 'Materials',
    '20': 'Industrials',
    '25': 'Consumer Discretionary',
    '30': 'Consumer Staples',
    '35': 'Health Care',
    '40': 'Financials',
    '45': 'Information Technology',
    '50': 'Communication Services',
    '55': 'Utilities'
}

# Map to the GIC Sectors
df['sector'] = df['GIC Sectors'].astype(str).map(industry_mapping)

# Calculate the average and median ESG scores for each sector
sector_stats = df.groupby('sector')['ESG Score'].agg(['mean', 'median']).round(2)
sector_stats
```

### **18. Pairwise Comparison of ESG Scores and Financial Performance**

For each ESG Grade (i.e., `ESG Grade`, `E Grade`, `S Grade`, `G Grade`), calculate the average and median values of the following financial variables:

- `Book Value Per Share`
- `Cash and Short-Term Investments`

```python
def calculate_statistics(group_column):
    """ Function to calculate average and median for each ESG category """
    # I) First financial variable = 'Book Value Per Share'
    # II) Second financial variable = "Cash and Short-Term Investments"
    df_stats = df.groupby(group_column)[
        ['Book Value Per Share', 'Cash and Short-Term Investments']].agg(['mean', 'median'])
    return df_stats.round(2) # We round the results with 2 digits

def plot_metric_trend(stats_df):
    """ Function to plot the mean and median trends for the calculated statistics """
    # We reverse the order of the rows to start with lower grades 
    stats_df = stats_df.iloc[::-1]

    fig, axes = plt.subplots(2, 1, figsize=(5.5, 4), sharex=True)

    # I) Plot for 'Book Value Per Share'
    sns.lineplot(data=stats_df["Book Value Per Share"], x=stats_df.index, y='mean', ax=axes[0],
                 label='Mean', marker='o', linestyle='-', sort=False)
    sns.lineplot(data=stats_df["Book Value Per Share"], x=stats_df.index, y='median', ax=axes[0],
                 label='Median', marker='o', linestyle='--', sort=False)

    axes[0].set_title('Book Value Per Share per Grade', fontsize=10)
    axes[0].set_xlabel(stats_df.index.name)
    axes[0].set_ylabel('Book Value \n Per Share')
    axes[0].grid(True)
    axes[0].legend(bbox_to_anchor=(1.02, 0.75), loc='upper left')

    # II) Plot for 'Cash and Short-Term Investments'
    sns.lineplot(data=stats_df["Cash and Short-Term Investments"], x=stats_df.index, y='mean', ax=axes[1],
                 label='Mean', marker='o', linestyle='-', sort=False)
    sns.lineplot(data=stats_df["Cash and Short-Term Investments"], x=stats_df.index, y='median', ax=axes[1],
                 label='Median', marker='o', linestyle='--', sort=False)

    axes[1].set_title(r'Cash and Short-Term Investments per Grade', fontsize=10)
    axes[1].set_xlabel('ESG Grade')
    axes[1].set_ylabel('Cash and Short-Term \n Investments')
    axes[1].grid(True)
    axes[1].legend(bbox_to_anchor=(1.02, 0.75), loc='upper left')

    plt.tight_layout()
    plt.show()

# Calculating the average and median for ESG/E/S/G Grades
esg_stats = calculate_statistics('ESG Grade')
display(esg_stats)
plot_metric_trend(esg_stats)

e_stats = calculate_statistics('E Grade')
display(e_stats)
plot_metric_trend(e_stats)

s_stats = calculate_statistics('S Grade')
display(s_stats)
plot_metric_trend(s_stats)

g_stats = calculate_statistics('G Grade')
display(g_stats)
plot_metric_trend(g_stats)
```

### **21. Top 5 Companies with Highest ESG Improvement**

For each company, calculate the year-over-year change in the `ESG Score`.

Identify the top 5 companies with the largest positive change in ESG score between two consecutive years. Provide the ticker, company name, and the ESG score change for each.


```python
# Sort the DataFrame by Ticker and Year
df = df.sort_values(by=['Ticker', 'Year'])

# Calculate the year-over-year change in ESG Score for each company
df['ESG Score Change'] = df.groupby('Ticker')['ESG Score'].diff()

# Identify the top 5 companies with the largest positive ESG score change
top_5_positive_change = df.nlargest(5, 'ESG Score Change')[['Ticker', 'Company Name', 'Year', 'ESG Score Change']]

top_5_positive_change.reset_index(drop=True)
```

```python
selected_tickers = top_5_positive_change["Ticker"].values

df_filtered = df[df['Ticker'].isin(selected_tickers)][['Year', 'Ticker', 'Company Name', 'ESG Score']]

plt.figure(figsize=(7, 3))

sns.lineplot(data=df_filtered, x='Year', y='ESG Score', hue='Company Name', marker='o')

plt.title('ESG Score Trends Over Time', fontsize=10)
plt.xlabel('Year')
plt.ylabel('ESG Score')
plt.grid(True)
plt.legend(title='Company', bbox_to_anchor=(1.02, 0.75), loc='upper left')
plt.tight_layout()
plt.show()
```
